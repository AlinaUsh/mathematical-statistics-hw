```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2, norm
```

```python
x = np.array(range(10, 5000, 50))
y = []
gamma = 0.9
print(f' gamma = {gamma}')

for n in x:
    m = 0
    for iter in range(0, 100):
        sample = np.random.normal(0., 4., n)
        x_sq_sum = (sample ** 2).sum()
        l_bound = x_sq_sum / chi2.ppf((1 + gamma) / 2, n)
        r_bound = x_sq_sum / chi2.ppf((1 - gamma) / 2, n)
        m += r_bound - l_bound
    y.append(m / 100)

y = np.array(y)
plt.grid()
plt.plot(x, y)
plt.xlabel('sample size')
plt.ylabel('interval length')
```

```python
x = np.array(range(10, 5000, 50))
y = []
gamma = 0.9
print(f' gamma = {gamma}')

for n in x:
    m = 0 
    for iter in range(0, 100):
        sample = np.random.normal(0., 4., n)
        x_mean_sq = (sample.mean()) ** 2
        l_bound = n * x_mean_sq / (norm.ppf((3 + gamma) / 4) ** 2)
        r_bound = n * x_mean_sq / (norm.ppf((3 - gamma) / 4) ** 2)
        m += r_bound - l_bound
    y.append(m / 100)

y = np.array(y)
plt.grid()
plt.plot(x, y)
plt.xlabel('sample size')
plt.ylabel('interval length')
```
