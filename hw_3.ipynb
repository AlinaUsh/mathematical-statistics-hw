```python
import random
import numpy as np
import matplotlib.pyplot as plt
import heapq
```

а)

Зависимость времени ожидания $x$-ого человека в очереди от различных $\alpha$ при фиксированном числе окошек $n = 100$.


```python
n = 100
p = n * 2
alpha_list = [random.random() * 10 for i in range(0, 5)]
alpha_list.sort()
print(f'Список параметров для эксперимента: {alpha_list}')

for alpha in alpha_list:
    waiting = np.zeros(p)
    for i in range(1000):
        time = np.random.exponential(alpha, p)
        h = list(time[:n])
        heapq.heapify(h)
        for j in range(n, p):
            m = heapq.heappop(h)
            waiting[j] += m
    waiting /= 1000
    plt.grid()
    plt.plot(np.arange(p), waiting, label=f'alpha = {alpha}')
    plt.legend()
```

Время ожидания $x$-ого человека в очереди при различных $n$ и фиксированном $\alpha$.

```python
alpha = 3
for n in [16, 50, 100, 256]:
    p = n * 2
    waiting = np.zeros(p)
    for i in range(1000):
        time = np.random.exponential(alpha, p)
        h = list(time[:n])
        heapq.heapify(h)
        for j in range(n, p):
            m = heapq.heappop(h)
            waiting[j] += m
            heapq.heappush(h, m + time[j])
    waiting /= 1000
    plt.grid()
    plt.plot(np.arange(p), waiting, label=f'n = {n}')
    plt.legend()
```
